```python
import numpy as np
import matplotlib.pyplot as plt

# Vector as arrow
v1 = np.array([10,5])          # magnitude = 5
v2 = np.array([5 , 4]) 
print("Vector v:", v1)
print("Magnitude:", np.linalg.norm(v1))
print("Vector v:", v2)
print("Magnitude:", np.linalg.norm(v2))

# Plot1
plt.figure(figsize=(6,6))
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='red')
plt.xlim(-1,10); plt.ylim(-1,10); plt.grid()
plt.title("Vector [3,4]"); plt.show()
# Plot2
plt.figure(figsize=(6,6))
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red')
plt.xlim(-1,10); plt.ylim(-1,10); plt.grid()
plt.title("Vector [3,4]"); plt.show()
```
